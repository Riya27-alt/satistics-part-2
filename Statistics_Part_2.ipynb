Statistics theory qustions
1. What is hypothesis testing in statistics?
Hypothesis testing is a method to make decisions or inferences about a population based on sample data. It helps determine whether an observed effect is statistically significant or just due to chance.

2. What is the null hypothesis and how does it differ from the alternative hypothesis?
Null Hypothesis (H‚ÇÄ): Assumes no effect or no difference (e.g., "There is no difference between group A and B").

Alternative Hypothesis (H‚ÇÅ or Ha): Assumes there is an effect or a difference.

3.  What is the significance level in hypothesis testing, and why is it important?
The significance level (Œ±) is the probability of rejecting the null hypothesis when it's actually true.

Common values: 0.05, 0.01.

It sets the threshold for what is considered statistically significant.

4. What does a P-value represent in hypothesis testing?
The P-value is the probability of getting the observed result (or more extreme) if the null hypothesis is true.

5. How do you interpret the P-value in hypothesis testing?
If P ‚â§ Œ±, reject the null hypothesis (result is significant).

If P > Œ±, fail to reject the null (result is not significant).

6. What are Type I and Type II errors?
Type I Error (False Positive): Rejecting a true null hypothesis.

Type II Error (False Negative): Failing to reject a false null hypothesis.

7. Difference between one-tailed and two-tailed tests:
One-tailed test: Looks for an effect in one direction.

Two-tailed test: Tests for any difference (either direction).

8. What is the Z-test, and when is it used?
Used when:

The population variance is known.

Sample size is large (n > 30).

Data is normally distributed.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
Used when:

The population variance is unknown.

Sample size is small (n < 30).

11. Difference between a Z-test and a T-test:
Z-test: Known population standard deviation.

T-test: Unknown population standard deviation.

12. What is the T-test, and how is it used?
Tests if the means of two groups are significantly different. Types include:

One-sample

Independent two-sample

Paired sample

13. Relationship between Z-test and T-test:
Both are used for comparing means, but Z-test uses the normal distribution while T-test uses the t-distribution (especially with smaller samples).

14. What is a confidence interval and how is it used?
It gives a range of values likely to contain the population parameter. A 95% confidence interval means we are 95% confident that the true value lies in this range.

15. What is the margin of error and how does it affect the confidence interval?
The margin of error is the range added/subtracted from the sample estimate to create the confidence interval. A larger margin = wider interval.

16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes' Theorem updates the probability of a hypothesis based on new evidence. It is widely used in diagnostic testing and decision making.

17. What is the Chi-square distribution, and when is it used?
Used for categorical data to test relationships or goodness of fit between observed and expected frequencies.

18. What is the Chi-square goodness of fit test, and how is it applied?
It tests if a sample matches an expected distribution. Example: testing if a die is fair.

19. What is the F-distribution, and when is it used?
Used in ANOVA and F-tests to compare variances across groups.

20. What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) compares the means of 3 or more groups. Assumptions:

Normal distribution

Equal variances

Independent samples

21. What are the different types of ANOVA tests?
One-way ANOVA: One independent variable.

Two-way ANOVA: Two independent variables.

Repeated Measures ANOVA: Same subjects measured multiple times.

22. What is the F-test, and how does it relate to hypothesis testing?
It compares variances. Used in ANOVA to check if group variances are significantly different.

```python
from scipy.stats import norm
import numpy as np

sample = np.random.normal(105, 10, 30)
population_mean = 100
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
np.random.seed(0)
data = np.random.normal(100, 15, 50)
mean = np.mean(data)
z = (mean - 100) / (15 / np.sqrt(50))
p = 2 * (1 - norm.cdf(abs(z)))

print("Simulated Mean:", mean)
print("Z-score:", z)
print("P-value:", p)
```

```python
def one_sample_z_test(sample, population_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - population_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = np.random.normal(102, 10, 40)
z, p = one_sample_z_test(sample, 100, 10)
print("Z-statistic:", z, "P-value:", p)
```

```python
def plot_type1_type2(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    null = norm.pdf(x, 0, 1)
    alt = norm.pdf(x, 1.5, 1)

    plt.plot(x, null, label='H0', color='blue')
    plt.plot(x, alt, label='H1', color='green')
    plt.fill_between(x, null, where=(x > norm.ppf(1 - alpha)), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, alt, where=(x < norm.ppf(1 - alpha)), color='orange', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

plot_type1_type2()
```

```python
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    sem = stats.sem(data)
    interval = stats.t.interval(confidence, len(data)-1, loc=mean, scale=sem)
    return interval

sample = np.random.normal(50, 8, 40)
ci = confidence_interval(sample)
print("Confidence Interval:", ci)
```

```python
def margin_of_error(data, confidence=0.95):
    sem = stats.sem(data)
    moe = sem * stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return moe

sample = np.random.normal(60, 12, 30)
print("Margin of Error:", margin_of_error(sample))
```

```python
def bayes_theorem(p_A, p_B_given_A, p_B_given_not_A):
    p_not_A = 1 - p_A
    p_B = p_B_given_A * p_A + p_B_given_not_A * p_not_A
    p_A_given_B = (p_B_given_A * p_A) / p_B
    return p_A_given_B

# Example: Disease test
print("P(Disease | Positive Test):", bayes_theorem(0.01, 0.99, 0.05))
```
